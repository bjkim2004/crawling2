```python
from datetime import datetime
import pandas as pd
import matplotlib as plt
from matplotlib import font_manager, rc
import os
import numpy as np
```

```python
def set_font(font_file='malgun.ttf'):
    font_dir = 'c:/Windows/Fonts/'
    font_name = font_manager.FontProperties(fname=os.path.join(font_dir, font_file)).get_name()
    rc('font', family=font_name)
```

```python
set_font()
```

```python
df_exchange = pd.read_csv('DEXKOUS.csv', header=1, 
    names=['DATE', 'DEXKOUS'], skipinitialspace=True, index_col=0)
```

```python
df_exchange.info(), df_exchange.describe()
```

```python
# df_exchange[df_exchange.DEXKOUS=='.']['DEXKOUS'] = np.NaN
df_exchange.DEXKOUS = df_exchange.DEXKOUS.replace(".", np.NaN)
df_exchange.DEXKOUS = df_exchange.DEXKOUS.fillna(method='ffill')
df_exchange.DEXKOUS = df_exchange.DEXKOUS.astype('float')
```

```python
df_exchange.info(), df_exchange.describe()
```

```python
df_exchange.DEXKOUS.plot(figsize = (17, 4), style='r')
```
